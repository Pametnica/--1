#  RPY (Roll Pitch Yaw) агли

```python
Една роботска рака може да ја оствари саканата ориентација во просторот и со следните 3 последователни ротации околу
оските на подвижниот координатен систем придружен кон неа (n,o,a), во дадениот редослед:
    - ротација за агол фи околу а оска- наречен агол на скршнување
    - ротација за агол пси мало околу о-оската, наречен агол на скршнување
    - ротација за агол ета околу н- оската наречен агол на валање
```
